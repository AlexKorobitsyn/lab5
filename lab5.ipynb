```python
import matplotlib.pyplot as plt
import numpy as np


def analitic_answer(x : np.array):
    return 2*np.square(x-1)
X1 = np.array([i*0.0333333333 for i in range(0, 31, 1)])
X2 = np.array([i*0.05 for i in range(0, 21, 1)])
X3 = np.array([i*0.1 for i in range(0, 11, 1)])
plt.title("Аналитическое решение")

plt.plot(X1, analitic_answer(X1), label=r"N = 30")
# plt.plot(X2, analitic_answer(X2), label=r"N = 20")
# plt.plot(X3, analitic_answer(X3), label=r"N = 10")
plt.legend(fontsize=12, loc="best")
plt.grid(True)
plt.savefig("Analitic111.png")
plt.show()
```

```python
def f(x, y):
    return 2*y/(x-1)


def euler_explicit(h, y_0, b = 1, x_0 = 0 ):
    X = []
    Y = []
    while(abs(x_0 - b - h) > 10e-4):
        X.append(x_0)
        Y.append(y_0)
        y_1 = y_0 + f(x_0, y_0) * h
        x_0 += h
        y_0 = y_1
    return X, Y
X1, Y1 = euler_explicit(0.0333333333 , 2)
print(X1, Y1)
X2, Y2 = euler_explicit(0.05 , 2)
print(X2, Y2)
X3, Y3 = euler_explicit(0.1 , 2)
print(X3, Y3)

plt.plot(X1, Y1, label=r"N = 30")
plt.plot(X2, Y2, label=r"N = 20")
plt.plot(X3, Y3, label=r"N = 10")
plt.xlim([0.4, 0.6])  # Устанавливаем пределы оси X
plt.ylim([0.25, 0.75]) 
plt.plot(X1, analitic_answer(np.array(X1)), label=r"Точный метод")
plt.title("Явный метод Эйлера")
plt.legend(fontsize=12, loc="best")
plt.grid(True)
plt.savefig("euler111.png")
plt.show()
```

```python
def koshi(h, y_0, x_0=0, b=1):
    X = []
    Y = []
    while(abs(x_0 - b - h) > 10e-4):
        X.append(x_0)
        Y.append(y_0)
        y_05 = y_0 + h/2*f(x_0, y_0)
        y_1 = y_0 + h*f(x_0+h/2, y_05)
        x_0 += h
        y_0= y_1
    return X, Y
X1, Y1 = koshi(0.0333333333 , 2)
print(X1, Y1)
X2, Y2 = koshi(0.05 , 2)
print(X2, Y2)
X3, Y3 = koshi(0.1 , 2)
print(X3, Y3)

plt.plot(X1, Y1, label=r"N = 30")
plt.plot(X2, Y2, label=r"N = 20")
plt.plot(X3, Y3, label=r"N = 10")
plt.xlim([0.475, 0.525])  # Устанавливаем пределы оси X
plt.ylim([0.5, 0.525]) 
plt.plot(X3, analitic_answer(np.array(X3)), label=r"Точный метод")
plt.title("Метод Коши")
plt.legend(fontsize=12, loc="best")
plt.grid(True)
plt.savefig("koshi111.png")
plt.show()
```

```python
def method_Adams_Multhon(h, y_0, x_0=0, b=1):
    X = []
    Y = []
    X.append(x_0)
    Y.append(y_0)
    x_1 = x_0+h
    x_2 = x_1+h
    y_1 = y_0 + 2*y_0*h/(x_0-1) + (h**2)/2*(2*y_0/((x_0-1)**2))
    X.append(x_1)
    Y.append(y_1)
    while(abs(x_2 - b-h) > 10e-4):
        y_2 = y_1*6*(x_2-1)/(6*x_2-6-5*h)+h*y_1*8*(x_2-1)/((x_1-1)*(6*x_2-6-5*h))-h*y_0*(x_2-1)/(2*(6*x_2-6-5*h)*(x_0-1))
        X.append(x_2)
        Y.append(y_2)
        y_0 = y_1
        y_1 = y_2
        x_0 += h
        x_1 += h
        x_2 += h
    return X, Y

X1, Y1 = method_Adams_Multhon(0.0333333333 , 2)
X2, Y2 = method_Adams_Multhon(0.05, 2)
X3, Y3 = method_Adams_Multhon(0.1, 2)
plt.figure(figsize=(15,15))
plt.plot(X1, Y1, label=r"N = 30")
plt.plot(X2, Y2, label=r"N = 20")
plt.plot(X3, Y3, label=r"N = 10")
# plt.xlim([0.0, 0.050])  # Устанавливаем пределы оси X
# plt.ylim([1.75, 2]) 
plt.plot(X1, analitic_answer(np.array(X1)), label=r"Точный метод")
plt.title("Метод Адамса-Мултона 3 порядок")
plt.legend(fontsize=12, loc="best")
plt.grid(True)
plt.savefig("adams111.png")
plt.show()
```
